### A Problem with Presidents

```python
#importing libraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data as dataframe
df = pd.read_csv("U.S. Presidents Birth and Death Information - Sheet1.csv")
df.head()
```

```python
df.drop(df.tail(1).index,inplace=True)
df
```

```python
df.dtypes
```

```python
# Converting 'BIRTH DATE' and 'DEATH DATE' to datetime format
df['BIRTH DATE'] = pd.to_datetime(df['BIRTH DATE'])
df['DEATH DATE'] = pd.to_datetime(df['DEATH DATE'])

#Adding column year_of_birth
df['year_of_birth'] = df['BIRTH DATE'].dt.year
df.head()
```

```python
# Calculating lived_years
df['lived_years'] = df.apply(lambda row: (row['DEATH DATE'].year - row['year_of_birth']) if pd.notnull(row['DEATH DATE']) else (datetime.now().year - row['year_of_birth']), axis=1)

# Calculate lived_months 
df['lived_months'] = df.apply(lambda row: (row['DEATH DATE'].year - row['year_of_birth']) * 12 + row['DEATH DATE'].month - row['BIRTH DATE'].month if pd.notnull(row['DEATH DATE']) else (datetime.now().year - row['year_of_birth']) * 12 + datetime.now().month - row['BIRTH DATE'].month, axis=1)

# Calculate lived_days 
df['lived_days'] = df.apply(lambda row: (row['DEATH DATE'] - row['BIRTH DATE']).days if pd.notnull(row['DEATH DATE']) else (datetime.now() - row['BIRTH DATE']).days, axis=1)

df.head()
```

```python
# Finding the top 10 longest-lived presidents
top_10_longest_lived = df.nlargest(10, 'lived_days')
top_10_longest_lived
top_10_longest_lived.to_csv("top_10_longest_lived.csv", index=False)
```

```python
# Finding the 10 shortest-lived presidents
top_10_shortest_lived = df.nsmallest(10, 'lived_days')
top_10_shortest_lived
top_10_shortest_lived.to_csv("top_10_shortest_lived.csv", index=False)
```

```python
# Describing the data
df.describe()
```

```python
# Calculate descriptive statistics

statistics = {
    'Mean': df['lived_days'].mean(),
    'Weighted Average': np.average(df['lived_days'], weights= df['lived_years']),
    'Median': df['lived_days'].median(),
    'Mode': df['lived_days'].mode().values,
    'Max': df['lived_days'].max(),
    'Min': df['lived_days'].min(),
    'Standard Deviation': df['lived_days'].std()
}

# Display the statistics
statistics_df = pd.DataFrame.from_dict(statistics, orient='index', columns=['Value'])
statistics_df.index.name = 'Statistic'
statistics_df
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, gridspec_kw={'height_ratios': [3, 1]})

# Box plot
ax1.boxplot(df['lived_days'].dropna(),  patch_artist=True, vert=False)

# Highlight minimum, maximum, and standard deviation
ax1.axvline(statistics['Min'], color='red', linestyle='--', label='Min')
ax1.axvline(statistics['Max'], color='green', linestyle='--', label='Max')
ax1.axvline(statistics['Mean'] - statistics['Standard Deviation'], color='purple', linestyle='--', label='1 SD')
ax1.axvline(statistics['Mean'] + statistics['Standard Deviation'], color='purple', linestyle='--')
ax1.legend()

ax1.set_ylabel('Distribution')
ax1.set_title('Distribution of Lived Days')

# Histogram
sns.histplot(df['lived_days'].dropna(), ax=ax2, kde=True)

ax2.set_xlabel('Lived Days')
ax2.set_ylabel('Density')

plt.tight_layout()
plt.show()
```
